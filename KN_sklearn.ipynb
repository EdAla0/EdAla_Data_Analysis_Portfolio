```python
import matplotlib.pyplot as plt
from keras.datasets import mnist
from sklearn.datasets import make_blobs, make_classification, make_moons, make_circles, load_digits
#from sklearn.linear_model import Perception
from sklearn.neighbors import KNeighborsClassifier
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.inspection import DecisionBoundaryDisplay
X1,y = make_moons(n_samples = 200, noise = 0.15)
X, Xp, y, yp= train_test_split(X1, y, test_size = 0.2, stratify = y)
model = KNeighborsClassifier(n_neighbors = 5)
model.fit(X,y)
predicciones = model.predict(Xp)
print(predicciones)
print(accuracy_score(yp, predicciones))
DecisionBoundaryDisplay.from_estimator(model, X1, response_method = 'predict')
plt.scatter(X[:,0], X[:,1], c = y, edgecolor = 'gray')
plt.scatter(Xp[:,0], Xp[:,1], c = yp, edgecolor = 'red')
plt.show()
```
